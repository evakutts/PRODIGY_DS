Generating a bar chart to depict the population distribution of top 10 countries in the year 2020.

```python
countries = data['Country Name']
population_2022 = data['2020']


sorted_data = data.sort_values(by='2020', ascending=False)


top_n = 10
top_countries = sorted_data['Country Name'].head(top_n)


plt.figure(figsize=(12, 8))
plt.bar(top_countries, sorted_data['2020'].head(top_n), color='green')
plt.xlabel('Country')
plt.ylabel('Population in 2020')
plt.title(f'Distribution of Population in 2020 - Top {top_n} Countries')
plt.xticks(rotation=45)  
plt.show()
```

BARCHART: Creating a bar chart to visualize selected countries populations in the year 1990.

```python
# Extracting required columns 
countries = data['Country Name']
population_1990 = data['1990']

# List of countries 
selected_countries = ['China', 'India', 'Russia', 'Japan', 'Canada', 'Australia', 'Germany', 'France', 'Samoa', 'South Korea']

# Filtering the DataFrame for the selected countries
selected_data = data[data['Country Name'].isin(selected_countries)]

# Sorting
sorted_data = selected_data.sort_values(by='1990', ascending=False)

# Creating a bar chart
plt.figure(figsize=(12, 8))
plt.bar(sorted_data['Country Name'], sorted_data['1990'], color='purple')
plt.xlabel('Country')
plt.ylabel('Population in 1990')
plt.title(f'Distribution of Population in 1990')
plt.xticks(rotation=45)
plt.show()
```

Creating a bar chart to visualize selected countries populations in the year 2010.

```python
# Extracting required columns 
countries = data['Country Name']
population_2010 = data['2010']

# List of countries 
selected_countries = ['China', 'India', 'Russia', 'Japan', 'Canada', 'Australia', 'Germany', 'France', 'Samoa', 'South Korea']

# Filtering the DataFrame for the selected countries
selected_data = data[data['Country Name'].isin(selected_countries)]

# Sorting
sorted_data = selected_data.sort_values(by='2010', ascending=False)

# Creating a bar chart
plt.figure(figsize=(12, 8))
plt.bar(sorted_data['Country Name'], sorted_data['2010'], color='orange')
plt.xlabel('Country')
plt.ylabel('Population in 2010')
plt.title(f'Distribution of Population in 2010')
plt.xticks(rotation=45)
plt.show()
```

HISTOGRAM:Histogram for the Population Change in Afghanistan from 1960 to 2022:

```python
import matplotlib.pyplot as plt

afghanistan_population_change = data[data['Country Name'] == 'Afghanistan'].iloc[:, 5:-1].diff(axis=1).iloc[:, -1]

plt.figure(figsize=(10, 6))
plt.hist(afghanistan_population_change, bins=15, color='orange', edgecolor='black')
plt.xlabel('Population Change (1960-2022)')
plt.ylabel('Frequency')
plt.title('Population Change Distribution in Afghanistan (1960-2022)')
plt.show()
```
